```python
import numpy as np
import pandas as pd
```

```python
list_ = [1,2,3,4,5,6,9,10,12,55,60]
```

```python
list_
```

```python
Q1 = np.percentile(list_,25)
```

```python
Q1
```

```python
Q3 = np.percentile(list_,75)
```

```python
Q3
```

```python
IQR =  Q3-Q1
```

```python
IQR
```

```python
low = Q1 - 1.5*IQR
```

```python
upper = Q3 + 1.5*IQR
```

```python
outliers =[]
```

```python
list_
```

```python
for i in list_:
    if((i>upper)or(i>low)):
        outliers.append(i)
```

```python
outliers
```

```python
list_ = [i for i in list_ if i not in outliers]
list_
```
